```python
import cv2

img = cv2.imread('earrings.jpg')
points = []
```

```python
def click_event(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        points.append((x, y))
        cv2.circle(img, (x, y), 3, (0, 255, 0), -1)
        cv2.putText(img, f'({x},{y})', (x+5, y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1)
        cv2.imshow('Earrings - Click corners (press ESC when done)', img)
        print(f"Point {len(points)}: ({x}, {y})")
```

```python
cv2.imshow('Earrings - Click corners (press ESC when done)', img)
cv2.setMouseCallback('Earrings - Click corners (press ESC when done)', click_event)

while True:
    key = cv2.waitKey(1) & 0xFF
    if key == 27: 
        break

cv2.destroyAllWindows()
```

```python
if len(points) >= 2:
    x1, y1 = points[0]
    x2, y2 = points[1]
    width_pixels = abs(x2 - x1)
    height_pixels = abs(y2 - y1) if len(points) >= 4 else width_pixels
    
    f_mm = 8
    D_mm = 720
    pixel_size_mm = 0.0022

    D_prime = 1 / ((1/f_mm) - (1/D_mm))
    m = D_prime / D_mm
    width_mm = (width_pixels * pixel_size_mm) / m
    height_mm = (height_pixels * pixel_size_mm) / m
    
    print(f"\n=== RESULTS ===")
    print(f"Width in pixels: {width_pixels}")
    print(f"Height in pixels: {height_pixels}")
    print(f"Actual earring width: {width_mm:.2f} mm ({width_mm/10:.2f} cm)")
    print(f"Actual earring height: {height_mm:.2f} mm ({height_mm/10:.2f} cm)")
else:
    print("Not enough points selected.")
```
